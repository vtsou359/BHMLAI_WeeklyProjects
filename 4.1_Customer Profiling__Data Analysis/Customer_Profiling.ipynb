## Analysis 1: Education Level and Promotional Campaign Response

In this analysis, we explore the relationship between customers' education level and their likelihood to accept offers in the last promotional campaign. We'll visualize how different education levels correlate with campaign response rates.

### Question: Is there a relationship between the type of education and the acceptance of the last promotional campaign?


````python
# Group data by education level and sum the response values to see total acceptances per education level
# The commented line below shows the data transformation approach
#df[["Education","Response"]].groupby("Education").sum().sort_values(by = "Response").reset_index()

fig, ax = plt.subplots(figsize=(10, 6))

sns.set_theme(style="whitegrid")

sns.barplot(data = df[["Education","Response"]].groupby("Education").sum().sort_values(by = "Response").reset_index()
            ,x="Education", y = 'Response',
            palette = ['#191D88', '#1450A3', '#337CCF', '#FFC436']
           )

ax.bar_label(ax.containers[0],
             fmt='%.f', label_type='edge')


plt.title('Number of customers who accepted the offer in the last campaign grouped by Education level')
plt.xlabel("Education")
plt.ylabel("Number of positive responses")

plt.savefig('Plot1.png', dpi =200)
````

## Analysis 3: Family Composition and Purchase Channels

In this analysis, we examine how the number of children and teenagers in a household affects customers' purchasing behavior across different channels (web, catalog, store, deals). We'll investigate whether families with children have different shopping preferences.

### Question: Are customers with children more likely to purchase products online? Which purchase channels are preferred by families with different numbers of children?


````python
# Extract and group data by number of children and teenagers in the household
# The commented line below was used for initial data exploration
#df.info()

temp2 = df[['Kidhome',
        'Teenhome', 
        'NumDealsPurchases',
        'NumWebPurchases', 
        'NumCatalogPurchases',
        'NumStorePurchases']].groupby(['Kidhome', 'Teenhome']).sum().reset_index()#.plot(kind = 'bar')

temp2
````

### Data Transformation for Purchase Channel Analysis

To analyze purchase patterns by family composition, we need to reshape our data. We'll:
1. Convert the data from wide to long format using the `melt` function
2. Calculate the total number of children and teenagers in each household
3. Rename the purchase types for better readability
4. Group the data by family composition and purchase type


````python
# Reshape data from wide to long format using melt
# This transforms the different purchase types into a single column with corresponding values
df3 = pd.melt(temp2, id_vars=['Kidhome', 'Teenhome'], 
              value_vars=['NumDealsPurchases',
                         'NumWebPurchases',
                         'NumCatalogPurchases',
                         'NumStorePurchases'
                         ],
              var_name='Type of Purchase', 
              value_name='Number of Purchase')

df3['Number of kids and teens in families'] = df3['Kidhome'] +df3['Teenhome']

df3['Type of Purchase'] = df3['Type of Purchase'].replace('NumDealsPurchases', 'Deals')
df3['Type of Purchase'] = df3['Type of Purchase'].replace('NumWebPurchases', 'Web')
df3['Type of Purchase'] = df3['Type of Purchase'].replace('NumCatalogPurchases', 'Catalog')
df3['Type of Purchase'] = df3['Type of Purchase'].replace('NumStorePurchases', 'Store')

for_sns = df3[['Type of Purchase','Number of Purchase','Number of kids and teens in families']].groupby(['Number of kids and teens in families', 'Type of Purchase']).sum().reset_index()

for_sns
````

### Visualization of Purchase Patterns by Family Composition

Now we'll create a bar chart to visualize how the number of children and teenagers in a household affects purchasing behavior across different channels. This will help us understand if families with children prefer online shopping or other purchase methods.


````python
# Create a bar plot to visualize purchase patterns by family composition
# The commented questions below guided our analysis



fig, ax = plt.subplots(figsize=(15, 7))

plt.title('Number of purchases grouped by number of total teenagers/kids in households (per type of purchase)')


sns.set_style("whitegrid")
#sns.set_style("dark")
sns.barplot(data = for_sns, 
            x = 'Number of kids and teens in families',
            y = 'Number of Purchase',
            hue = 'Type of Purchase',
            palette=['#BFEAF5', '#91C8E4', '#749BC2', '#4682A9'],
            edgecolor='none',
            ax=ax )

ax.bar_label(ax.containers[0],
             fmt='%.f', label_type='edge')
ax.bar_label(ax.containers[1],
             fmt='%.f', label_type='edge')
ax.bar_label(ax.containers[2],
             fmt='%.f', label_type='edge')
ax.bar_label(ax.containers[3],
             fmt='%.f', label_type='edge')

plt.xlabel("Total Number of kids/teenagers in households")
plt.ylabel("Number of purchases")


plt.savefig('Plot3.png', dpi =200)
````

## Analysis 4: Product Types and Customer Complaints

In this analysis, we investigate which types of product purchases are associated with customer complaints. We'll examine the maximum amount spent on different product categories by customers who have filed complaints.

### Question: What kinds of purchases led to customer complaints?


````python
# Filter data to include only customers who have complained
# The commented lines below were used for initial data exploration
#df.info()
#df['Complain'].value_counts()

df4 = df.query('Complain == 1')[[ 'MntMeatProducts', 
                                       'MntFishProducts', 
                                       'MntSweetProducts', 
                                       'MntGoldProds', 
                                       'MntFruits',
                                       'MntWines',
                                       'Complain']].groupby('Complain').max().reset_index()
df4
````

### Data Transformation for Customer Complaints Analysis

To analyze which product categories are associated with customer complaints, we need to:
1. Reshape the data to have product types as a single variable
2. Rename the product categories for better readability
3. Prepare the data for visualization


````python
# Reshape data from wide to long format using melt
# This transforms the different product types into a single column with corresponding maximum amounts
for_sns2 = pd.melt(df4, id_vars=['Complain'], 
                          value_vars=['MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'MntFruits', 'MntWines'],
                          var_name='Type of Product', 
                          value_name='Max Amount').set_index('Complain')


for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntMeatProducts', 'Meat')
for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntFishProducts', 'Fish')
for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntGoldProds', 'Gold Products')
for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntFruits', 'Fruits')
for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntWines', 'Wines')
for_sns2['Type of Product'] = for_sns2['Type of Product'].replace('MntSweetProducts', 'Sweet Products')


for_sns2
````

### Visualization of Product Types Associated with Customer Complaints

Now we'll create a bar chart to visualize the maximum amount spent on different product categories by customers who have filed complaints. This will help us identify which product categories might be associated with higher complaint rates.


````python
# Create a bar plot to visualize maximum spending by product category for customers with complaints
# The commented questions below guided our analysis

fig, ax = plt.subplots(figsize=(15, 6))

plt.title('Maximum amount spent for every type of product in which customers complained')
plt.xlabel("Product Type")
plt.ylabel("Maximum Amount spent")


ax =  sns.barplot(data = for_sns2.sort_values('Max Amount'),
                     x = 'Type of Product',
                     y = 'Max Amount',
                     edgecolor='none',
                     ax=ax,
                     palette=['#DDE6ED', '#9DB2BF', '#526D82', '#27374D', '#57C5B6', '#159895'])


plt.savefig('Plot4.png', dpi =200)
````
